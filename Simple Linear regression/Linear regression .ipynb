```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
#let's plot and see how our data looks and how many stochastic error do we have 
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
# now you can see here that our data is not linear it is sort of linear

# now let's apply the algorithms and make the model so that whenever we give the cgpa we can understand what is the package of the person 

X = df.iloc[:,0:1]  
Y = df.iloc[:,-1]   
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)  # means 20% data we take for testing
```

```python
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)   # here we are training by .fit()
```

```python
X_test
```

```python
Y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package in lpa')
```

```python
#y=mx+b
#to find slop or m 
m=lr.coef_
```

```python
b = lr.intercept_
```

```python
b
```

```python
# y = mx + b

# now let's c if our predicted values was correct or not

y1 = m*8.58 + -0.8961119222429144
```
