```python
import pandas as pd
import numpy as np 
from sklearn.datasets import fetch_california_housing
```

```python
data=fetch_california_housing()
```

```python
# Convert to DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df['Target'] = data.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import seaborn as sns

sns.histplot(df['MedInc'],kde=True)
```

```python
#The distribution is positively skewed 
#lets see if we apply log transformation what will happen
```

```python
print(df['MedInc'].mean())
print(df['MedInc'].median())
print(df['MedInc'].mode())
```

```python
df['MedInc']=np.log1p(df['MedInc'])
```

```python

sns.histplot(df['MedInc'],kde=True)
```

```python
#still its heavy tailed 
kurtosis_values = df['MedInc'].kurt()
print(kurtosis_values)
```

```python
sns.boxplot(x='MedInc',data=df)
```

```python
sns.pairplot(data=df,hue='Target')
```
