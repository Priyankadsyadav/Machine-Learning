```python
import numpy as np #for making numpy as array
import pandas as pd #used for making datafarme
import matplotlib.pyplot as plt #used for making graphs
import seaborn as sns 
import sklearn.datasets #to get datasets and machine learning algorithms
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics #for evaluation
```

```python
house_price_dataset=sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset) #array is stored in target label
```

```python
# Loading the dataset to pandas dataframe
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head(5)
```

```python
#adding the target(price) column to the DataFrame
house_price_dataframe['price']=house_price_dataset.target
```

```python
house_price_dataframe.head(5)
```

```python
#checking the number of rows and columns in dataframe
house_price_dataframe.shape
```

```python
#checking for missing values
house_price_dataframe.isnull().sum()
```

```python
#luckily there are no missing values 

# Let's see statistical measures of the datset
house_price_dataframe.describe()
```

```python
#understanding the correlation between various features in the dataset 
#1. Positive correlation 
#2. Negative correlation 
```

```python
correlation = house_price_dataframe.corr()
```

```python
#constructing heatmap ,as heatmap gives understanding for correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1F',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
#splitting the data and Target
X=house_price_dataframe.drop(['price'],axis=1)
Y=house_price_dataframe['price']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model training using XGboost Regressor
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,Y_train)
```

```python
#Evaluation
#Calculating R squared error
#before that we do prediction on training data 
#accuracy for prediction on training data 
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R squared error

score_1 = metrics.r2_score(Y_train,training_data_prediction)   #basically finds the variance between Y_train and training_data_prediction

#Mean absolute error

score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction) #basically finds the difference between Y_train and training_data_prediction
```

```python
print("R squared error: ", score_1)  
print("Mean Absolute error: ", score_2)
```

```python
# here R squared error is pretty small means model is performing good
```

```python
#lets plot and see the error values
# Visualising the actual and predicted values

plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted Price")
plt.show()
```

```python
# Prediction on test data
test_data_prediction=model.predict(X_test)
print(test_data_prediction)
```

```python
#R squared error

score_1 = metrics.r2_score(Y_test,test_data_prediction)   #basically finds the variance between Y_train and training_data_prediction

#Mean absolute error

score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction) #basically finds the difference between Y_train and training_data_prediction
```

```python
print("R squared error: ", score_1)  
print("Mean Absolute error: ", score_2)
```

```python
#lets plot and see the error values
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted Price")
plt.show()
```
