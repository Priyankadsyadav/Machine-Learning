#### 2. Loading pipeline

```python
from transformers import pipeline
translation_pipeline = pipeline('translation_en_to_de', device=0)
```

```python
import torch
print(torch.cuda.is_available())  # Should return True if GPU is set up correctly
```

```python
results = translation_pipeline('I love ice cream')
```

```python
results[0]['translation_text']
```

#### 3. Create Gradio Function and Interface

```python
def translate_transformers(from_text):
    results = translation_pipeline(from_text)
    return results[0]['translation_text']
```

```python
translate_transformers('My name is Priyanka')
```

```python
import gradio as gr

interface = gr.Interface(
    fn=translate_transformers, 
    inputs=gr.Textbox(lines=2, placeholder='Text to translate'),  # Updated to gr.Textbox
    outputs=gr.Textbox()  # Updated to gr.Textbox for outputs as well
)

# Launch the interface
interface.launch()
```
